```python
# 18. Class: thuoc tinh - phuong thuc(function-def)
class person:
    def __init__(self, name, age):
        self.name = name # self = person
        self.age = age
    def infor(self):
        print("Name: ", self.name)
        print("Age: ", self.age)
#-------------------------------------
p1 = person("Nguyen Anh Tuan", 35)
p1.infor()
p1.age = 50
p1.infor()
p1.name = 'Nguyen Tuan Anh'
p1.infor()
```

```python
# 18. class: Giai phong memory
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def pr_infor(self):
        print("Name: ", self.name)
        print("Age: ", self.age)
#-----
p2 = person("Nguyen Van A", 36)
p2.pr_infor()
del p2 # giai phong memory
p2.pr_infor()
```

```python
# 18. class: Su ke thua, bo sung them phuong thuc, thuoc tinh
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def infor(self):
        print("Name: ", self.name)
        print("Age: ", self.age)

class student(person):
    def __init__(self, name, age, year):
        super().__init__(name, age)
        self.graduation = year
    def infor(self):
        super().infor()
        print("Graduation year: ", self.graduation)
#----------------
p = student("Nguyen Van A", 35, 2012)
p.infor()
a = "A"
b = 36
c = 2015
p = student(a, b, c)
p.infor()
```

```python
# 18. class: Hinh chu nhat
class rectange:
    def __init__(self, dai, rong):
        self.dai = dai
        self.rong = rong
    def chuvi(self):
        chu_vi = (self.dai+self.rong)*2
        return chu_vi
    def dientich(self):
        dien_tich = self.dai*self.rong
        return dien_tich

p = rectange(30,20)
print("Chu vi: ", p.chuvi())
print("Dien tich: ", p.dientich())
```

```python
# cv2.imresize() and plot
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Loading the image
image_path = r'/home/anton/Downloads/Lenna_(test_image).png'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)
bigger = cv2.resize(image, (1050, 1610))

stretch_near = cv2.resize(image, (780, 540),
			interpolation = cv2.INTER_NEAREST)

Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image, half, bigger, stretch_near]
count = len(images)

for i in range(count):
	plt.subplot(2, 2, i + 1)
	plt.title(Titles[i])
	plt.imshow(images[i])

plt.show()
```
